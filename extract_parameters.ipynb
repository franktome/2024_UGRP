```python
with open('content/vehicle_depth.json', 'r') as file:
    vehicle_depth = json.load(file)
    #print(vehicle_depth)
    #1. frame num 2. x, y, z
print(vehicle_depth['vehicle0'])
```

```python
import matplotlib.pyplot as plt
def depth_visualization(vehicle):
    how_far = vehicle_depth[vehicle]
    print()
        
    plt.figure(figsize=(10, 6))  # 그래프 크기 설정
    plt.plot(how_far, marker='o', linestyle='-', color='b')  # 선 그래프
    plt.title('Depth Visualization')  # 그래프 제목
    plt.xlabel('Index')  # x축 라벨
    plt.ylabel('Value')  # y축 라벨
    plt.grid(True)  # 그리드 추가
    plt.show()  # 그래프 출력
```

```python
vehicles = vehicle_depth.keys()
for vehicle in vehicles:
    print(f'{vehicle}: {len(vehicle_depth[vehicle])}')
    print(vehicle_depth[vehicle])
    depth_visualization(vehicle)
```
