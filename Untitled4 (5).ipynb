```python
import numpy as np
import cv2
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from skimage import io

# Load the npz files
binary_data = np.load('content/binary_image.npz')
embedding_data = np.load('content/embedding_image.npz')

# Extract the arrays
binary_image = binary_data['binary_image']
embedding_image_original = embedding_data['embedding_image']

# Ensure the embedding image has 3 channels
embedding_image = embedding_image_original[:, :, 0:3]

# Create a masked embedding image
masked_embedding_image = np.zeros_like(embedding_image)
mask = binary_image > 0
masked_embedding_image[mask] = embedding_image[mask]

# Extract the height and width of the image
h, w, _ = masked_embedding_image.shape

# Create a mask for non-black pixels (assuming black pixels are [0, 0, 0])
mask = np.any(masked_embedding_image != [0, 0, 0], axis=-1)

# Extract the RGB values and their corresponding coordinates
pixels = masked_embedding_image[mask].reshape(-1, 3)
coords = np.column_stack(np.nonzero(mask))

# Combine RGB values with coordinates
features = np.concatenate([pixels, coords], axis=1)

# Perform DBSCAN clustering
eps = 30  # Epsilon parameter for DBSCAN (this may require tuning)
min_samples = 100  # Minimum samples for a core point in DBSCAN
db = DBSCAN(eps=eps, min_samples=min_samples).fit(features)
labels = db.labels_

# Create an empty image for the clustered result
clustered_img = np.zeros_like(masked_embedding_image)

# Get the unique labels
unique_labels = np.unique(labels)
n_clusters = len(unique_labels) - (1 if -1 in labels else 0)  # Exclude noise if present

# Generate random colors for each cluster
colors = np.random.randint(0, 255, size=(n_clusters, 3))

# Assign cluster colors to the original image based on labels
cluster_centers = {label: [] for label in unique_labels if label != -1}

for label, (x, y) in zip(labels, coords):
    if label != -1:  # Ignore noise
        clustered_img[x, y] = colors[label % n_clusters]
        cluster_centers[label].append((x, y))

# Calculate the centroid of each cluster
centroids = {label: np.mean(points, axis=0).astype(int) for label, points in cluster_centers.items()}

# Plot the results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Masked Embedding Image')
plt.imshow(cv2.cvtColor(masked_embedding_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Clustered Image with Spatial Information')
plt.imshow(cv2.cvtColor(clustered_img, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Annotate each cluster with its index
for label, (cx, cy) in centroids.items():
    plt.text(cy, cx, str(label), color='white', fontsize=12, ha='center', va='center')

plt.show()
```

```python
# resize bounding box
original_size = (720, 1280)
new_size = (256, 512)

height_scale = new_size[0] / original_size[0]
width_scale = new_size[1] / original_size[1]

def resize_vehicle_coordinates(vehicle_coords, height_scale, width_scale):
    resized_coords = {}
    for key, coords in vehicle_coords.items():
        x1, y1, x2, y2 = coords
        x1_new = x1 * width_scale
        y1_new = y1 * height_scale
        x2_new = x2 * width_scale
        y2_new = y2 * height_scale
        resized_coords[key] = np.array([x1_new, y1_new, x2_new, y2_new], dtype=np.float32)
    return resized_coords

resized_points = resize_vehicle_coordinates(points, height_scale, width_scale)
```

```python
# 각 vehicle bounding box의 밑변 중점 좌표 계산
midpoints = {}
for vehicle, coords in resized_points.items():
    x1, y1, x2, y2 = coords
    midpoint = ((x1 + x2) / 2, y2)
    midpoints[vehicle] = midpoint
```

```python
# 클러스터 좌표 추출 함수
def extract_clusters(clustered_img, labels):
    unique_labels = np.unique(labels)
    clusters = []

    for label in unique_labels:
        if label == -1:  # Ignore noise
            continue

        # 클러스터 좌표 추출
        cluster_coords = np.column_stack(np.where(labels == label))
        clusters.append(cluster_coords)

    return clusters

# 클러스터 좌표 추출
clusters = extract_clusters(clustered_img, labels)
```

```python
# 각 vehicle 밑변 중점과 가장 가까운 클러스터 idx 계산
def find_nearest_cluster(midpoint, clusters):
    min_dist = float('inf')
    nearest_cluster_idx = -1
    for idx, cluster in enumerate(clusters):
        dists = np.linalg.norm(cluster - midpoint, axis=1)
        min_cluster_dist = np.min(dists)
        if min_cluster_dist < min_dist:
            min_dist = min_cluster_dist
            nearest_cluster_idx = idx
    return nearest_cluster_idx
```

```python
# plot vehicle bounding box on clustered image

lane_image = clustered_img

# plot vehicle bounding box
for key, (x1, y1, x2, y2) in resized_points.items():
    cv2.rectangle(lane_image, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
    cv2.putText(lane_image, key.replace("vehicle", ''), (int(x1), int(y1)-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1, cv2.LINE_AA)

# 밑변 중점 좌표 그리기
for vehicle, midpoint in midpoints.items():
    cv2.circle(lane_image, (int(midpoint[0]), int(midpoint[1])), 5, (255, 255, 255), -1)

# 중점과 가장 가까운 클러스터를 찾고, 해당 클러스터가 왼쪽에 있는지 오른쪽에 있는지 판단
for vehicle, midpoint in midpoints.items():
    nearest_cluster_idx = find_nearest_cluster(midpoint, clusters)
    nearest_cluster = clusters[nearest_cluster_idx]

    # 클러스터의 중심을 계산하여 왼쪽인지 오른쪽인지 판단
    cluster_center = np.mean(nearest_cluster, axis=0)
    side = "left" if cluster_center[0] < midpoint[0] else "right"

    print(f"{vehicle}({midpoint[0]}, {midpoint[1]}): nearest cluster is {nearest_cluster_idx}, {side} side")

plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(lane_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Annotate each cluster with its index
for label, (cx, cy) in centroids.items():
    plt.text(cy, cx, str(label), color='white', fontsize=12, ha='center', va='center')
    
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to process an image and return the blank image with the extended line
def process_image(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding
    _, binary_image = cv2.threshold(gray_image, 1, 255, cv2.THRESH_BINARY)

    # Detect edges using Canny
    edges = cv2.Canny(binary_image, 50, 150, apertureSize=3)

    # Detect lines using the probabilistic Hough transform
    lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=15, minLineLength=50, maxLineGap=10)

    # Create a blank image to draw the extended line
    blank_image = np.zeros_like(image)

    # If lines are detected, process the first detected line
    if lines is not None and len(lines) > 0:
        print(f"Number of lines detected: {len(lines)}")
        # Take the first detected line
        x1, y1, x2, y2 = lines[0][0]

        # Extend the line
        extension_length = 500  # Length to extend on each side
        angle = np.arctan2(y2 - y1, x2 - x1)
        x1_ext = int(x1 - extension_length * np.cos(angle))
        y1_ext = int(y1 - extension_length * np.sin(angle))
        x2_ext = int(x2 + extension_length * np.cos(angle))
        y2_ext = int(y2 + extension_length * np.sin(angle))

        # Draw the extended line on the blank image
        cv2.line(blank_image, (x1_ext, y1_ext), (x2_ext, y2_ext), (255, 255, 255), 2)

    return blank_image, (x1_ext, y1_ext), (x2_ext, y2_ext)

def count(image, y, x):
    result_count = 0
    for i in range(x):
        if image[y][i][0] == 0 and image[y][i+1][0] != 0:
            result_count += 1
    return result_count

def perpendicular_distance(point, line_point1, line_point2):
    x0, y0 = point
    x1, y1 = line_point1
    x2, y2 = line_point2
    
    numerator = abs((y2 - y1)*x0 - (x2 - x1)*y0 + x2*y1 - y2*x1)
    denominator = np.sqrt((y2 - y1)**2 + (x2 - x1)**2)
    
    distance = numerator / denominator
    return distance

def find_closest_index(midpoints, line_segments):
    closest_indices = {}
    for vehicle, midpoint in midpoints.items():
        closest_index = None
        min_distance = float('inf')
        for idx, (p1, p2) in enumerate(line_segments):
            distance = perpendicular_distance(midpoint, p1, p2)
            if distance < min_distance:
                min_distance = distance
                closest_index = idx
        closest_indices[vehicle] = closest_index
    return closest_indices

# Initialize a blank image to accumulate results
accumulated_image = np.zeros_like(cluster_imgs[0])
line_segments = []

# Process images from cluster_imgs[0] to cluster_imgs[4]
for i in range(5):
    processed_image, p1, p2 = process_image(cluster_imgs[i])
    accumulated_image = cv2.add(accumulated_image, processed_image)
    line_segments.append((p1, p2))

# Display the result
plt.imshow(cv2.cvtColor(accumulated_image, cv2.COLOR_BGR2RGB))
plt.title('Accumulated Extended Lines')
plt.axis('off')
plt.show()

c_num = count(accumulated_image, 200, 500)
print(f"Number of lines crossing y=200 within first 500 columns: {c_num}")

closest_indices = find_closest_index(midpoints, line_segments)
print("Closest indices for each vehicle:", closest_indices)
```
