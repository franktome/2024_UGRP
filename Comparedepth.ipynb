```python
# depth 추출
import numpy as np
import os

def load_npy_files(directory):
    npy_files = [f for f in os.listdir(directory) if f.endswith('metric_depth.npy')]
    data_list = []
    for file in npy_files:
        data = np.load(os.path.join(directory, file))
        data_list.append(data)
    return data_list

directory_path = 'content/object_depths'
depth_map_list = load_npy_files(directory_path)

# depth map 길이 (=프레임 개수)
print(len(depth_map_list))
```

```python
import matplotlib.pyplot as plt
import numpy as np


x_index = 540  # 특정 x 값 (열 번호)

# 해당 x 값의 모든 y 값 추출
depth_values = depth_map_list[0][:, x_index]  # 특정 열 추출 (행 전체)

# y 값 범위 (행 번호)
y_values = np.arange(len(depth_values))

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(y_values, depth_values,  color='b', label=f"Depth at x={x_index}")
plt.title(f"Depth Values at x={x_index}", fontsize=16)
plt.xlabel("y (row index)", fontsize=14)
plt.ylabel("Depth[m]", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

# 이미지 경로
img_path = 'content/tracker/000000.png'

# 이미지 로드
img = mpimg.imread(img_path)

# 이미지 표시
plt.imshow(img)
plt.axis('off')  # 축 숨기기
plt.title("Loaded Image")

# x_index에 해당하는 세로선 그리기 (이미지에)
x_index = 970  # 특정 x 값 (열 번호)
plt.axvline(x=x_index, color='r', linestyle='-', linewidth=2, label=f"x={x_index}")  # 빨간색 세로선

# depth_map_list에서 x_index에 대한 depth 값 추출
depth_values = depth_map_list[0][:, x_index]  # 특정 열 추출 (행 전체)
y_values = np.arange(len(depth_values))  # y 값 범위 (행 번호)

# 그래프 그리기 (이미지 위에 추가)
plt.figure(figsize=(8, 6))
plt.plot(y_values, depth_values, color='b', label=f"Depth at x={x_index}")
plt.title(f"Depth Values at x={x_index}", fontsize=16)
plt.xlabel("y (row index)", fontsize=14)
plt.ylabel("Depth[m]", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg

# 이미지 경로 설정
img_path = 'content/tracker/000000.png'

# 이미지 로드
img = mpimg.imread(img_path)

# 이미지 표시
plt.imshow(img)
plt.axis('off')  # 축 숨기기
plt.title("Loaded Image")

# y_index에 해당하는 가로선 그리기 (이미지에)
y_index = 600  # 특정 y 값 (행 번호)
plt.axhline(y=y_index, color='r', linestyle='-', linewidth=2, label=f"y={y_index}")  # 빨간색 가로선

# depth_map_list에서 y_index에 대한 depth 값 추출
depth_values = depth_map_list[0][y_index, :]  # 특정 행 추출 (열 전체)
x_values = np.arange(len(depth_values))  # x 값 범위 (열 번호)

# 그래프 그리기 (이미지 위에 추가)
plt.figure(figsize=(8, 6))
plt.plot(x_values, depth_values, color='b', label=f"Depth at y={y_index}")
plt.title(f"Depth Values at y={y_index}", fontsize=16)
plt.xlabel("x (column index)", fontsize=14)
plt.ylabel("Depth[m]", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
